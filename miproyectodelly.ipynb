```python
mediana=df['Review Rating'].median()
print(mediana)
```

```python
#imputar valores nulos
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))  
```

```python
#identificar los nulos
df.isnull().sum()
```

```python
#cambiamos a minuscula y llenamos los espacios en blanco con guion abajo a los titulos o campos
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace('  ','_')
```

```python
df.columns
```

```python
df=df.rename(columns= {'purchase_amount_(usd)' :' purchase_amount'})
```

```python
df.columns
```

```python
import pandas as pd
bins = [0,17,35,60,100] 
labels = ['menor de edad', 'joven','adulto','sinior']

#etiqueta
df ['grupo_edad']=pd.cut(df['age'],bins=bins,labels=labels)
```

```python
#mostrar las columnas
df[['age','grupo_edad']].head(10)
```

```python
df
```

```python

# Crear diccionario para convertir la frecuencia de compra en días

frequency_mapping = {
    'Weekly': 7,
    'Bi-Weekly': 14,
    'Fortnightly': 14,
    'Monthly': 30,
    'Quarterly': 90,
    'Every 3 Months': 90,
    'Annually': 365
}

# Crear nueva columna con la frecuencia en días

df['frequency_of_purchases_days'] = df['frequency of purchases'].map(frequency_mapping)
```

```python
df.columns
```

```python
df[['frequency of purchases', 'frequency_of_purchases_days']].head(10)
```

```python
# Cambiar los nombres de las columnas
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('(', '', regex=False)
    .str.replace(')', '', regex=False)
)

# Verificar
print(df.columns)
```

```python
print(df.columns)
```

```python
# Revisar las columnas de descuento y código promocional
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
#eliminamos una de ellas 
df=df.drop('promo_code_used',axis=1)
```
